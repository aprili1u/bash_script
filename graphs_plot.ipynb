```python
import numpy as np  
import seaborn as sns  
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = [i for i in range (1,10000)]
data = pd.read_csv("data_100interactions.csv",header = None) 
i = 0
fig = plt.figure(figsize=(12, 7))
for i in range (len(data)):
    dt = data.iloc[i]
    if (dt[0] == 0 or dt[0] == 1 or dt[0] == 2):
        i+=1
        if (dt[0] == 0):
            plt.plot(list(dt[1:]),color= 'deepskyblue')
        if (dt[0] == 1):
            plt.plot(list(dt[1:]),color= 'blue')
        if (dt[0] == 2):
            plt.plot(list(dt[1:]),color= 'gold')
            
plt.xlabel('Generations')
plt.ylabel('Mean memory')
plt.title("10 individuals 100 interactions randomly picked for each individual, everyone is connected")
plt.show()
```
