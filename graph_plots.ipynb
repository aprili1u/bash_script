```python
import numpy as np
import seaborn as sns
import csv
import pandas as pd
import matplotlib.pyplot as plt

def plot(n1,n2,n3,line,title):
    X = [i for i in range (1,10000)]
    data = pd.read_csv('results_cleaned.csv', header=None, error_bad_lines=False)
    i = 0
    fig = plt.figure(figsize=(12, 7))
    for i in range (len(data)):
        dt = data.iloc[i]
        if (dt[0] == n1 or dt[0] == n2 or dt[0] == n3):
            i+=1
            if (dt[0] == n1 and dt[2] == line):
                plt.plot(list(dt[3:]),color= 'deepskyblue')
            if (dt[0] == n2 and dt[2] == line):
                plt.plot(list(dt[3:]),color= 'blue')
            if (dt[0] == n3 and dt[2] == line):
                plt.plot(list(dt[1:]),color= 'gold')

    plt.xlabel('Generations')
    plt.ylabel('Mean memory')
    plt.title(title)
    plt.show()
```

```python
plot(1,2,1,'Mean_Memory','title')
```
